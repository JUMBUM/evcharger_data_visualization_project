```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/전기차_충전소_데이터_0602_10시04분.csv')
df = df.drop_duplicates('주소(구)')
df
```

```python
df1 = df.groupby('주소(구)', as_index = False).sum()
df1
```

```python
df2 = df.drop_duplicates(['충전소명'])
df2 = df2.groupby('주소(구)', as_index = False).count()
df2
```

```python
df3 = pd.merge(df1, df2, on = '주소(구)', how = 'inner')
df3
```

```python
df3['충족율'] = (df3['충전소명'] / df3['ele_num_x'])*100
df3 = df3.sort_values(by = '충족율', ascending = True)
df3 = df3[['충족율', '주소(구)']]
df3
```

```python
import seaborn as sns
fig = plt.figure(figsize=(25,10))
sns.barplot(x='주소(구)',y='충족율',data=df3)
plt.show()
```
